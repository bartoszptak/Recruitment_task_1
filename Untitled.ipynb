```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
graduates = pd.read_csv('data/graduates.csv')
score_board = pd.read_csv('data/score_board.csv')
```

```python
graduates.describe()
```

```python
score_board['accepted'].value_counts()
```

```python
df = pd.merge(score_board, graduates, on='id', how='left')
df[['graduated']] = df[['graduated']].fillna(value=False)

df = df.dropna(how='any')
df = df.drop(columns=['id', 'year', 'accepted'])

df['graduated'] = df['graduated'].replace(False, 0)

df.head(6)
```

```python
#df['gpa_score'] = df['gpa']*df['score']
#df['maths_score'] = df['maths_exam']*df['gpa']
df.head(6)
```

```python
df.describe()
```

```python
print(df[df['graduated']==True]['score'].min())
print(df[df['graduated']==True]['score'].mean())
print(df[df['graduated']==True]['score'].max())
```

```python
social_values = df['social_activity'].value_counts(normalize=True).sort_index()
social_values_grad = df[df['graduated']==True]['social_activity'].value_counts(normalize=True).sort_index()

social_plot = pd.DataFrame({'all': social_values, 'graduated':social_values_grad})
social_plot.plot.bar(rot=0)

plt.show()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns='graduated')
y = df[['graduated']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
gbm = xgb.XGBClassifier()
gbm
```

```python
gbm.fit(X_train, y_train.values.ravel())
```

```python
predictions = gbm.predict(X_test)
print('Accuracy: {:.2f}%'.format(accuracy_score(y_test, predictions)*100))
```

```python
impors = gbm.feature_importances_
labels = X_train.columns

for i in sorted(zip(labels, impors), key=lambda a: a[1], reverse=True):
    print(i)
```
